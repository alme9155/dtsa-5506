```python
import pandas as pd

# Load the data
tracking = pd.read_csv('/kaggle/input/nfl-big-data-bowl-2025/tracking_week1.csv')
plays = pd.read_csv('/kaggle/input/nfl-big-data-bowl-2025/plays.csv')

# Preview the structure
print(tracking.columns)
print(plays.columns)

# Example: Find one play to study
sample_play = plays.sample(1, random_state=42)
gameId = sample_play['gameId'].values[0]
playId = sample_play['playId'].values[0]

print(f"Selected gameId: {gameId}, playId: {playId}")
print(sample_play[['playDescription', 'offenseFormation', 'personnelO']])

# Filter the tracking data for this play
play_tracking = tracking[(tracking['gameId'] == gameId) & (tracking['playId'] == playId)]

# Sort by frame to analyze frame-by-frame movement
play_tracking = play_tracking.sort_values(by=['frameId', 'nflId'])

# See player movement in first few frames
print(play_tracking.head())

# Optional: visualize players on field (basic)
import matplotlib.pyplot as plt

def plot_frame(df, frame_id):
    frame = df[df['frameId'] == frame_id]
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_xlim(0, 120)
    ax.set_ylim(0, 53.3)
    for _, row in frame.iterrows():
        ax.scatter(row['x'], row['y'], c='blue' if row['team'] == 'home' else 'red', label=row['team'])
        ax.text(row['x'], row['y'], row['jerseyNumber'], fontsize=8, ha='center')
    ax.set_title(f"Frame {frame_id}")
    plt.xlabel('X (yards)')
    plt.ylabel('Y (width)')
    plt.show()

# Plot the first frame of the play
plot_frame(play_tracking, frame_id=play_tracking['frameId'].min())
```
